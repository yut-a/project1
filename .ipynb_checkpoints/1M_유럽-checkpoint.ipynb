### 데이터 불러오고 이전 데이터 추정

```python
# 매크로 데이터 불러오기
import pandas as pd

macro = pd.read_csv("macro.csv", index_col = 0)
macro["date"] = pd.to_datetime(macro["date"])
macro.head()
```

```python
# ETF 데이터
import FinanceDataReader as fdr

euro_etf = fdr.DataReader("195930", "2010-01-01")
euro_etf = euro_etf.reset_index()
euro_etf.columns = ["date", "open", "high", "low", "close", "volume", "change"]
euro_etf.head()
```

```python
# EURO STOXX50 데이터

euro = fdr.DataReader("STOXX50", "2010-01-01")
euro = euro.reset_index()
euro.columns = ["date", "close", "open", "high", "low", "volume", "change"]
euro["high_change"] = (euro["high"] / euro["close"]) - 1
euro["low_change"] = (euro["low"] / euro["close"]) - 1
euro.head()
```

```python
# 2014년 4월 30일 이전 데이터 추정
use_euro = euro[euro["date"] <= "2014-04-30"].sort_values(by = "date", ascending = False).reset_index(drop = True)

euro_list = [10110]
for i in range(0, len(use_euro)):
    estimate = euro_list[-1] / (1 + use_euro["change"][i])
    euro_list.append(estimate)

euro_list = euro_list[:695]

low_list = []
high_list = []
for j in range(1, len(euro_list)):
    low_estimation = euro_list[j] * (1 + use_euro["low_change"][j])
    high_estimation = euro_list[j] * (1 + use_euro["high_change"][j])

    low_list.append(low_estimation)
    high_list.append(high_estimation)
```

```python
# 추정 데이터와 ETF 데이터 합치기
df_estimation = pd.DataFrame()
df_estimation["date"] = use_euro["date"][1:]
df_estimation["high"] = high_list
df_estimation["low"] = low_list
df_estimation["close"] = euro_list[1:]

df_estimation = df_estimation.sort_values(by = "date", ascending = True).reset_index(drop = True)

euro_etf = pd.concat([df_estimation, euro_etf], axis = 0).reset_index(drop = True)
euro_etf.head()
```

### 기술 지표 데이터 산출

```python
# true range

high_low = []
high_close = []
low_close = []

for i in range(0, len(euro_etf)):
    
    # |당일 고가 - 당일 저가|
    today_high = euro_etf["high"][i]
    today_low = euro_etf["low"][i]
    val_high_low = abs(today_high - today_low)
    high_low.append(val_high_low)
    
    # |당일 고가 - 전일 종가|
    if i - 1 == -1:
        recent_close = euro_etf["close"][i]
    
    else:
        recent_close = euro_etf["close"][i - 1]
        
    val_high_close = abs(today_high - recent_close)
    high_close.append(val_high_close)
    
    # |당일 저가 - 전일 종가|
    val_low_close = abs(today_low - recent_close)
    low_close.append(val_low_close)
    
true_range = []

for j in range(0, len(euro_etf)):
    tr_val = max(high_low[j], high_close[j], low_close[j])
    true_range.append(tr_val)
    
euro_etf["true_range"] = true_range
```

```python
# nomalized average true range
import numpy as np

natr = []

for i in range(0, len(euro_etf)):
    if i + 1 < 14:
        natr_val = np.nan
    
    else:
        natr_val = euro_etf["true_range"][i-13:i+1].sum() / euro_etf["close"][i] * 100
        
    natr.append(natr_val)
    
euro_etf["NATR"] = natr
```

```python
# moving average (5, 15, 30)

ma5 = []

for i in range(0, len(euro_etf)):
    if i + 1 < 5:
        ma5_val = np.nan
        
    else:
        ma5_val = euro_etf["close"][i-4:i+1].sum() / 5
        
    ma5.append(ma5_val)
    
ma15 = []

for j in range(0, len(euro_etf)):
    if j + 1 < 15:
        ma15_val = np.nan
        
    else:
        ma15_val = euro_etf["close"][j-14:j+1].sum() / 15
        
    ma15.append(ma15_val)
    
ma30 = []

for k in range(0, len(euro_etf)):
    if k + 1 < 30:
        ma30_val = np.nan
        
    else:
        ma30_val = euro_etf["close"][k-29:k+1].sum() / 30
        
    ma30.append(ma30_val)
    
euro_etf["MA5"] = ma5
euro_etf["MA15"] = ma15
euro_etf["MA30"] = ma30
```

```python
# bollinger band (upper, middle, lower band, band width)

middle_band = []

for i in range(0, len(euro_etf)):
    if i + 1 < 20:
        middle = np.nan
        
    else:
        middle = euro_etf["close"][i-19:i+1].sum() / 20
        
    middle_band.append(middle)
    
upper_band = []
lower_band = []

for j in range(0, len(euro_etf)):
    if j + 1 < 20:
        upper = np.nan
        lower = np.nan
        
    else:
        upper = middle_band[j] + (2 * euro_etf["close"][j-19:j+1].std())
        lower = middle_band[j] - (2 * euro_etf["close"][j-19:j+1].std())
        
    upper_band.append(upper)
    lower_band.append(lower)
    
euro_etf["upper_band"] = upper_band
euro_etf["middle_band"] = middle_band
euro_etf["lower_band"] = lower_band
euro_etf["band_width"] = euro_etf["upper_band"] - euro_etf["lower_band"]
```

```python
# on balance volume

obv = []

for i in range(0, len(euro_etf)):
    if i - 1 == -1:
        obv_val = euro_etf["volume"][i]
        
    else:
        if euro_etf["close"][i] > euro_etf["close"][i-1]:
            obv_val = obv_val + euro_etf["volume"][i]
        
        else:
            obv_val = obv_val - euro_etf["volume"][i]
            
    obv.append(obv_val)
    
euro_etf["OBV"] = obv
```

```python
# 5 days momentun, 20 days momentum (영업일 기준)
euro_etf["close_5lag"] = euro_etf["close"].shift(5)
euro_etf["momentum_5"] = (euro_etf["close"] - euro_etf["close_5lag"]) / euro_etf["close_5lag"]

euro_etf["close_20lag"] = euro_etf["close"].shift(20)
euro_etf["momentum_20"] = (euro_etf["close"] - euro_etf["close_20lag"]) / euro_etf["close_20lag"]
```

```python
# MACD

def fnMACD(m_Df, m_NumFast=12, m_NumSlow=26, m_NumSignal=9):
    test = pd.DataFrame()
    test['EMAFast'] = m_Df['close'].ewm( span = m_NumFast, min_periods = m_NumFast - 1).mean()
    test['EMASlow'] = m_Df['close'].ewm( span = m_NumSlow, min_periods = m_NumSlow - 1).mean()
    test['MACD'] = test['EMAFast'] - test['EMASlow']
    test['MACDSignal'] = test['MACD'].ewm( span = m_NumSignal, min_periods = m_NumSignal-1).mean()
    test['MACDDiff'] = test['MACD'] - test['MACDSignal']
    
    m_Df["MACD"] = test["MACD"]
    m_Df["MACDDiff"] = test["MACDDiff"]
    return m_Df

euro_etf = fnMACD(euro_etf)
```

```python
# RSI

def SMA(data, period = 30, columns = "close"):
    return data[columns].rolling(window = period).mean()

def EMA(data, period = 20, columns = "close"):
    return data[columns].ewm(span = period, adjust = False).mean()

def RSI(data, period = 14, columns = "close"):
    delta = data[columns].diff(1)
    delta = delta.dropna()
    
    up = delta.copy()
    down = delta.copy()
    
    up[up < 0] = 0
    down[down > 0] = 0
    data["up"] = up
    data["down"] = down
    
    AVG_Gain = SMA(data, period, columns = "up")
    AVG_Loss = abs(SMA(data, period, columns = "down"))
    RS = AVG_Gain / AVG_Loss
    
    RSI = 100.0 - (100.0 / (1.0 + RS))
    data["RSI"] = RSI
    
    data = data.drop(["up", "down"], axis = 1)
    
    return data

euro_etf = RSI(euro_etf, 14)
```

```python
# target 생성
def target(x):
    if x > 0:
        return 1
    
    elif x <= 0:
        return 0
    
euro_etf["close_-20lag"] = euro_etf["close"].shift(-20)
euro_etf["return"] = (euro_etf["close_-20lag"] - euro_etf["close"]) / euro_etf["close"]
euro_etf["grade"] = euro_etf["return"].apply(target)

euro_etf.head()
```

```python
# euro_etf과 macro date merge
euro_etf = pd.merge(euro_etf, macro, on = "date", how = "inner").reset_index(drop = True)
euro_etf.head()
```

```python
# 필요한 칼럼만 추출
euro50_etf = euro_etf[["date", "volume", "NATR", "momentum_5", "momentum_20", "wti", "diff", "ism", "vix", "jpy_krw", "usd_krw", "china_pmi", "grade"]]
euro50_etf.head()
```

```python
# volume 결측값을 0으로 채우기
euro50_etf["volume"] = euro50_etf["volume"].replace(np.nan, 0)
euro50_etf.head()
```

```python
# 결측치 제거
euro50_etf = euro50_etf.dropna(axis = 0).reset_index(drop = True)
euro50_etf.head()
```

### 머신러닝으로 테스트

```python
# train, test set 분리
train = euro50_etf[euro50_etf["date"] < "2018-01-01"].reset_index(drop = True)
test = euro50_etf[euro50_etf["date"] >= "2018-01-01"].reset_index(drop = True)

train.shape, test.shape
```

```python
# 필요없는 칼럼 제거
drop_col = ["date"]

train = train.drop(drop_col, axis = 1)
test = test.drop(drop_col, axis = 1)
```

```python
# baseline
print(train["grade"].value_counts(normalize = True))
print("-------")
print(test["grade"].value_counts(normalize = True))
```

```python
# feature, target 분리
target = "grade"

X_train = train.drop(target, axis = 1)
y_train = train[target]

X_test = test.drop(target, axis = 1)
y_test = test[target]
```

```python
# 하이퍼 파라미터를 최적화한 XGBClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

pipe = make_pipeline(
    MinMaxScaler(),
    XGBClassifier(random_state = 99)
)

params = {
    "xgbclassifier__scale_pos_weight" : [0.579374 / 0.420626, 1, 0.420626 / 0.579374],
    "xgbclassifier__max_depth" : [5, 10, 15, 20, 25],
    "xgbclassifier__min_child_weight" : [10, 15, 20, 25, 30],
    "xgbclassifier__learning_rate" : [0.001, 0.005, 0.01, 0.05],
    "xgbclassifier__subsample" : [0.3, 0.4, 0.5],
    "xgbclassifier__n_estimators" : [500, 1000],
    "xgbclassifier__gamma" : [0.25, 0.5, 0.7, 1.0]
}

clf = RandomizedSearchCV(
    pipe,
    params,
    n_iter = 100,
    cv = 5,
    scoring = "f1_weighted",
    verbose = 1,
    n_jobs = -1,
    random_state = 99
)

clf.fit(X_train, y_train)
```

```python
# 최적 하이퍼 파라미터 / CV score
print("최적 하이퍼파라미터: ", clf.best_params_, "\n")
print("CV accuracy score: ", clf.best_score_)
```

```python
# 최적 하이퍼 파라미터를 적용한 모델의 train, test set 정확도
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

pipe_opt = make_pipeline(
    MinMaxScaler(),
    XGBClassifier(n_estimators = 1000, max_depth = 5, min_child_weight = 30, subsample = 0.3,
                  learning_rate = 0.005, gamma = 0.3, scale_pos_weight = 0.40799 / 0.59201, random_state = 99, n_jobs = -1)
)

pipe_opt.fit(X_train, y_train)

print("Train set accuracy score: ", pipe_opt.score(X_train, y_train))
print("Test set accuracy score: ", pipe_opt.score(X_test, y_test))
```

```python
# confusion matrix
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt

flg, ax = plt.subplots()
pcm = plot_confusion_matrix(pipe_opt, X_test, y_test,
                           cmap = plt.cm.Blues,
                           ax = ax);

plt.title(f"Confusion matrix, n = {len(y_test)}", fontsize = 15)
plt.show()
```

```python
def vaa_score(df):
    df["close_1y_lag"] = df["close"].shift(252)
    df["close_6m_lag"] = df["close"].shift(121)
    df["close_3m_lag"] = df["close"].shift(60)
    
    df["momentum_1y"] = (df["close"] - df["close_1y_lag"]) - 1
    df["momentum_6m"] = (df["close"] - df["close_6m_lag"]) - 1
    df["momentum_3m"] = (df["close"] - df["close_3m_lag"]) - 1
    
    df["vaa_score"] = (12 * df["momentum_20"]) + (4 * df["momentum_3m"]) + (2 * df["momentum_6m"]) + (1 * df["momentum_1y"])
    
    return df
```

```python
euro_vaa = vaa_score(euro_etf)
```

```python
euro_result = euro_vaa[euro_vaa["date"] >= "2018-01-01"][["date", "change", "vaa_score"]].reset_index(drop = True)
euro_result = pd.merge(euro_result, euro50_etf["date"], on = "date", how = "inner")
euro_result["predict"] = pipe_opt.predict(X_test)
euro_result
```
