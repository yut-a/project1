```python
# 매크로 데이터 불러오기
import pandas as pd

macro = pd.read_csv("macro.csv", index_col = 0)
macro["date"] = pd.to_datetime(macro["date"])
macro.head()
```

```python
# ETF 데이터
import FinanceDataReader as fdr

kodex200 = fdr.DataReader("069500", "2010-01-01")
kodex200 = kodex200.reset_index()
kodex200.columns = ["date", "open", "high", "low", "close", "volume", "change"]
kodex200.head()
```

```python
# true range

high_low = []
high_close = []
low_close = []

for i in range(0, len(kodex200)):
    
    # |당일 고가 - 당일 저가|
    today_high = kodex200["high"][i]
    today_low = kodex200["low"][i]
    val_high_low = abs(today_high - today_low)
    high_low.append(val_high_low)
    
    # |당일 고가 - 전일 종가|
    if i - 1 == -1:
        recent_close = kodex200["close"][i]
    
    else:
        recent_close = kodex200["close"][i - 1]
        
    val_high_close = abs(today_high - recent_close)
    high_close.append(val_high_close)
    
    # |당일 저가 - 전일 종가|
    val_low_close = abs(today_low - recent_close)
    low_close.append(val_low_close)
    
true_range = []

for j in range(0, len(kodex200)):
    tr_val = max(high_low[j], high_close[j], low_close[j])
    true_range.append(tr_val)
    
kodex200["true_range"] = true_range
```

```python
# nomalized average true range
import numpy as np

natr = []

for i in range(0, len(kodex200)):
    if i + 1 < 14:
        natr_val = np.nan
    
    else:
        natr_val = kodex200["true_range"][i-13:i+1].sum() / kodex200["close"][i] * 100
        
    natr.append(natr_val)
    
kodex200["NATR"] = natr
```

```python
# moving average (5, 15, 30)

ma5 = []

for i in range(0, len(kodex200)):
    if i + 1 < 5:
        ma5_val = np.nan
        
    else:
        ma5_val = kodex200["close"][i-4:i+1].sum() / 5
        
    ma5.append(ma5_val)
    
ma15 = []

for j in range(0, len(kodex200)):
    if j + 1 < 15:
        ma15_val = np.nan
        
    else:
        ma15_val = kodex200["close"][j-14:j+1].sum() / 15
        
    ma15.append(ma15_val)
    
ma30 = []

for k in range(0, len(kodex200)):
    if k + 1 < 30:
        ma30_val = np.nan
        
    else:
        ma30_val = kodex200["close"][k-29:k+1].sum() / 30
        
    ma30.append(ma30_val)
    
kodex200["MA5"] = ma5
kodex200["MA15"] = ma15
kodex200["MA30"] = ma30
```

```python
# bollinger band (upper, middle, lower band, band width)

middle_band = []

for i in range(0, len(kodex200)):
    if i + 1 < 20:
        middle = np.nan
        
    else:
        middle = kodex200["close"][i-19:i+1].sum() / 20
        
    middle_band.append(middle)
    
upper_band = []
lower_band = []

for j in range(0, len(kodex200)):
    if j + 1 < 20:
        upper = np.nan
        lower = np.nan
        
    else:
        upper = middle_band[j] + (2 * kodex200["close"][j-19:j+1].std())
        lower = middle_band[j] - (2 * kodex200["close"][j-19:j+1].std())
        
    upper_band.append(upper)
    lower_band.append(lower)
    
kodex200["upper_band"] = upper_band
kodex200["middle_band"] = middle_band
kodex200["lower_band"] = lower_band
kodex200["band_width"] = kodex200["upper_band"] - kodex200["lower_band"]
```

```python
# on balance volume

obv = []

for i in range(0, len(kodex200)):
    if i - 1 == -1:
        obv_val = kodex200["volume"][i]
        
    else:
        if kodex200["close"][i] > kodex200["close"][i-1]:
            obv_val = obv_val + kodex200["volume"][i]
        
        else:
            obv_val = obv_val - kodex200["volume"][i]
            
    obv.append(obv_val)
    
kodex200["OBV"] = obv
```

```python
# 5 days momentun, 20 days momentum (영업일 기준)
kodex200["close_5lag"] = kodex200["close"].shift(5)
kodex200["momentum_5"] = (kodex200["close"] - kodex200["close_5lag"]) / kodex200["close_5lag"]

kodex200["close_20lag"] = kodex200["close"].shift(20)
kodex200["momentum_20"] = (kodex200["close"] - kodex200["close_20lag"]) / kodex200["close_20lag"]
```

```python
# target 생성
def target(x):
    if x > 0:
        return 1
    
    elif x <= 0:
        return 0
    
kodex200["close_-20lag"] = kodex200["close"].shift(-20)
kodex200["return"] = (kodex200["close_-20lag"] - kodex200["close"]) / kodex200["close"]
kodex200["grade"] = kodex200["return"].apply(target)

kodex200.head()
```

```python
# kodex200과 macro date merge
kodex200 = pd.merge(kodex200, macro, on = "date", how = "inner").reset_index(drop = True)
kodex200.head()
```

```python
# 필요한 칼럼만 추출
kodex200_etf = kodex200[["date", "volume", "NATR", "momentum_5", "momentum_20", "diff", "ism", "vix", "jpy_krw", "usd_krw", "china_pmi", "grade"]]
kodex200_etf.head()
```

```python
# 결측치 제거
kodex200_etf = kodex200_etf.dropna(axis = 0).reset_index(drop = True)
kodex200_etf.head()
```

### 머신러닝으로 테스트

```python
# train, test set 분리
train = kodex200_etf[kodex200_etf["date"] < "2018-01-01"].reset_index(drop = True)
test = kodex200_etf[kodex200_etf["date"] >= "2018-01-01"].reset_index(drop = True)

train.shape, test.shape
```

```python
# 필요없는 칼럼 제거
drop_col = ["date"]

train = train.drop(drop_col, axis = 1)
test = test.drop(drop_col, axis = 1)
```

```python
# baseline
print(train["grade"].value_counts(normalize = True))
print("-------")
print(test["grade"].value_counts(normalize = True))
```

```python
# feature, target 분리
target = "grade"

X_train = train.drop(target, axis = 1)
y_train = train[target]

X_test = test.drop(target, axis = 1)
y_test = test[target]
```

```python
# 하이퍼 파라미터를 최적화한 XGBClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

pipe = make_pipeline(
    MinMaxScaler(),
    XGBClassifier(random_state = 99)
)

params = {
    "xgbclassifier__scale_pos_weight" : [0.586298 / 0.413702, 1, 0.413702 / 0.586298],
    "xgbclassifier__max_depth" : [5, 10, 15, 20, 25],
    "xgbclassifier__min_child_weight" : [10, 15, 20, 25, 30],
    "xgbclassifier__learning_rate" : [0.001, 0.005, 0.01, 0.05],
    "xgbclassifier__subsample" : [0.3, 0.4, 0.5],
    "xgbclassifier__n_estimators" : [500, 1000],
    "xgbclassifier__gamma" : [0.25, 0.5, 0.7, 1.0]
}

clf = RandomizedSearchCV(
    pipe,
    params,
    n_iter = 100,
    cv = 5,
    scoring = "f1_weighted",
    verbose = 1,
    n_jobs = -1,
    random_state = 99
)

clf.fit(X_train, y_train)
```

```python
# 최적 하이퍼 파라미터 / CV score
print("최적 하이퍼파라미터: ", clf.best_params_, "\n")
print("CV accuracy score: ", clf.best_score_)
```

```python
# 최적 하이퍼 파라미터를 적용한 모델의 train, test set 정확도
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.decomposition import PCA

pipe_opt = make_pipeline(
    MinMaxScaler(),
    PCA(7),
    XGBClassifier(n_estimators = 1000, max_depth = 5, min_child_weight = 7, subsample = 0.3,
                  learning_rate = 0.0001, gamma = 0.3, scale_pos_weight = 0.413702 / 0.586298, random_state = 99, n_jobs = -1)
)

pipe_opt.fit(X_train, y_train)

print("Train set accuracy score: ", pipe_opt.score(X_train, y_train))
print("Test set accuracy score: ", pipe_opt.score(X_test, y_test))
```

```python
# confusion matrix
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt

flg, ax = plt.subplots()
pcm = plot_confusion_matrix(pipe_opt, X_test, y_test,
                           cmap = plt.cm.Blues,
                           ax = ax);

plt.title(f"Confusion matrix, n = {len(y_test)}", fontsize = 15)
plt.show()
```

```python
# VAA Score 산출
def vaa_score(df):
    df["close_1y_lag"] = df["close"].shift(252)
    df["close_6m_lag"] = df["close"].shift(121)
    df["close_3m_lag"] = df["close"].shift(60)
    
    df["momentum_1y"] = (df["close"] - df["close_1y_lag"]) - 1
    df["momentum_6m"] = (df["close"] - df["close_6m_lag"]) - 1
    df["momentum_3m"] = (df["close"] - df["close_3m_lag"]) - 1
    
    df["vaa_score"] = (12 * df["momentum_20"]) + (4 * df["momentum_3m"]) + (2 * df["momentum_6m"]) + (1 * df["momentum_1y"])
    
    return df
```

```python
kodex_vaa = vaa_score(kodex200)
```

```python
kospi_result = kodex_vaa[kodex_vaa["date"] >= "2018-01-01"][["date", "change", "vaa_score"]].reset_index(drop = True)
kospi_result = pd.merge(kospi_result, kodex200_etf["date"], on = "date", how = "inner")
kospi_result["predict"] = pipe_opt.predict(X_test)
kospi_result
```
