```python
# 데이터 불러오기
import pandas as pd

wti = pd.read_csv("wti.csv")[["Date", "Price"]]
vix = pd.read_csv("vix.csv")[["Date", "Price"]]
bond_3y = pd.read_csv("bond_3y.csv")[["Date", "Price"]]
bond_10y = pd.read_csv("bond_10y.csv")[["Date", "Price"]]
jpy_krw = pd.read_csv("jpy_krw.csv")[["Date", "Price"]]
usd_krw = pd.read_csv("usd_krw.csv")[["Date", "Price"]]
```

```python
# 전처리
def clean(df):
    df.rename(columns = {"Date" : "date", "Price" : "price"}, inplace = True)
    df["date"] = pd.to_datetime(df["date"])

    return df
```

```python
# clean 함수 적용
wti = clean(wti)
vix = clean(vix)
bond_3y = clean(bond_3y)
bond_10y = clean(bond_10y)
jpy_krw = clean(jpy_krw)
usd_krw = clean(usd_krw)
```

```python
# 장단기 금리차
bond = pd.merge(bond_3y, bond_10y, on = "date", how = "inner")

diff = pd.DataFrame()
diff["date"] = bond["date"]
diff["diff"] = bond["price_y"] - bond["price_x"]
```

```python
# pmi 데이터 불러오기
pmi = pd.read_csv("pmi.csv")
pmi = pmi.drop("Unnamed: 3", axis = 1)
pmi["date"] = pmi["date"].str.replace(".", ", 20")
pmi["date"] = pd.to_datetime(pmi["date"])
pmi.head()
```

```python
# 매크로 데이터 merge
macro = pd.merge(wti, vix, on = "date", how = "inner")
macro = pd.merge(macro, diff, on = "date", how = "inner")
macro = pd.merge(macro, jpy_krw, on = "date", how = "inner")
macro = pd.merge(macro, usd_krw, on = "date", how = "inner")
macro.columns = ["date", "wti", "vix", "diff", "jpy_krw", "usd_krw"]

macro = pd.merge(macro, pmi, on = "date", how = "outer").sort_values(by = "date", ascending = True).reset_index(drop = True)
macro = macro.fillna(method = "ffill")
macro
```

```python
# 결측치 제거
macro = macro.dropna(axis = 0).reset_index(drop = True)

# usd_krw 숫자형으로 변환
macro["usd_krw"] = macro["usd_krw"].str.replace(",", "")
macro["usd_krw"] = macro["usd_krw"].astype(np.float)
```

```python
macro.to_csv("macro.csv")
```
